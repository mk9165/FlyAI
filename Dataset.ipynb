## Text Summarization
https://aihub.or.kr/aihubdata/data/view.do?currMenu=115&topMenu=100&aihubDataSe=realm&dataSetSn=117

### csv

```python
import numpy as np
import pandas as pd
import pathlib
from tqdm import tqdm
```

```python
# 경로
train_dir = pathlib.Path('/content/drive/MyDrive/Colab Notebooks/FlyAI/project/summarization/한국어 대화 요약/Training')
val_dir = pathlib.Path('/content/drive/MyDrive/Colab Notebooks/FlyAI/project/summarization/한국어 대화 요약/Validation')
```

```python
# 파일 읽기
train_name = list(train_dir.glob('*.json'))
val_name = list(val_dir.glob('*.json'))
len(train_name), len(val_name)
```

```python
# train data
train_text = []
train_summary = []
for i in tqdm(range(len(train_name))):
  df = pd.read_json(train_name[i])
  for j in range(len(df)):
    train_summary.append(df['data'][j]['body']['summary'])

    text = []
    for k in range(len(df['data'][j]['body']['dialogue'])):
      text.append(df['data'][j]['body']['dialogue'][k]['utterance'])
    train_text.append(text)

# val data
val_text = []
val_summary = []
for i in tqdm(range(len(val_name))):
  df = pd.read_json(val_name[i])
  for j in range(len(df)):
    val_summary.append(df['data'][j]['body']['summary'])

    text = []
    for k in range(len(df['data'][j]['body']['dialogue'])):
      text.append(df['data'][j]['body']['dialogue'][k]['utterance'])
    val_text.append(text)
```

```python
len(train_text), len(train_summary), len(val_text), len(val_summary)
```

```python
train_df = pd.DataFrame({'text':train_text, 'summary':train_summary})
val_df = pd.DataFrame({'text':val_text, 'summary':val_summary})
```

```python
train_df['text'] = train_df['text'].apply(lambda x: ' '.join(x))
val_df['text'] = val_df['text'].apply(lambda x: ' '.join(x))
```

```python
train_df.head(3)
```

```python
train_df.to_csv('/content/drive/MyDrive/Colab Notebooks/FlyAI/project/summarization/summarization_train.csv', index=False)
val_df.to_csv('/content/drive/MyDrive/Colab Notebooks/FlyAI/project/summarization/summarization_val.csv', index=False)
```

## Sentiment Classification

```python
# train_data 
train_text = train_df['사람문장1'] + ' ' + train_df['시스템문장1'] + ' ' + train_df['사람문장2'] + ' ' + train_df['시스템문장2'] +  ' ' + train_df['사람문장3'] +  ' ' + train_df['시스템문장3']
train_label1 = train_df['감정_대분류']
train_label2 = train_df['감정_소분류']

# val_data 
val_text = val_df['사람문장1'] +  ' ' + val_df['시스템문장1'] +  ' ' + val_df['사람문장2'] +  ' ' + val_df['시스템문장2'] +  ' ' + val_df['사람문장3'] +  ' ' + val_df['시스템문장3']
val_label1 = val_df['감정_대분류']
val_label2 = val_df['감정_소분류']

train_df = pd.DataFrame({'text':train_text, 'label1':train_label1, 'label2': train_label2})
val_df = pd.DataFrame({'text':val_text, 'label1':val_label1, 'label2': val_label2})
```

```python
train_df['label'] = train_df['label1'].map({'분노':0, '슬픔':1, '불안':2, '상처':3, '당황':4, '기쁨':5})
val_df['label'] = val_df['label1'].map({'분노':0, '슬픔':1, '불안':2, '상처':3, '당황':4, '기쁨':5})
```

## VoiceFishing Classification

```python
import numpy as np
import pandas as pd
import pathlib
from tqdm import tqdm
```

```python
v_dir = pathlib.Path('/content/drive/MyDrive/Colab Notebooks/FlyAI/project/voicefishing')
```

```python
v_name = list(v_dir.glob('*.mp4'))
```

```python
from pydub import AudioSegment

sound = AudioSegment.from_file(v_name[5])
sound.export('wav_audio.wav', format='wav')
```

```python
import time
import azure.cognitiveservices.speech as speechsdk

def from_file():
    speech_config = speechsdk.SpeechConfig(subscription='d5b74499ac7d444f88d425157e2a1997', region='koreacentral')
    speech_config.speech_recognition_language='ko-kr'

    audio_config = speechsdk.audio.AudioConfig(filename='wav_audio.wav')
    speech_recognizer = speechsdk.SpeechRecognizer(speech_config=speech_config, audio_config=audio_config)

    done = False

    def stop_cb(evt):
        print('CLOSING on {}'.format(evt))
        nonlocal done
        done = True

    # Connect callbacks to the events fired by the speech recognizer
    speech_recognizer.recognizing.connect(lambda evt: print('RECOGNIZING: {}'.format(evt)))
   
    all_results = []
    def handle_final_result(evt):
        all_results.append(evt.result.text)

    speech_recognizer.recognized.connect(handle_final_result)

    speech_recognizer.recognized.connect(lambda evt: print('RECOGNIZED: {}'.format(evt)))
    speech_recognizer.session_started.connect(lambda evt: print('SESSION STARTED: {}'.format(evt)))
    speech_recognizer.session_stopped.connect(lambda evt: print('SESSION STOPPED {}'.format(evt)))
    speech_recognizer.canceled.connect(lambda evt: print('CANCELED {}'.format(evt)))
    # stop continuous recognition on either session stopped or canceled events
    speech_recognizer.session_stopped.connect(stop_cb)
    speech_recognizer.canceled.connect(stop_cb)

    # Start continuous speech recognition
    speech_recognizer.start_continuous_recognition()
    while not done:
        time.sleep(.5)

    speech_recognizer.stop_continuous_recognition()

    # if result.reason == speechsdk.ResultReason.RecognizedSpeech:
    #     print("Recognized: {}".format(result.text))
    # elif result.reason == speechsdk.ResultReason.NoMatch:
    #     print("No speech could be recognized: {}".format(result.no_match_details))
    # elif result.reason == speechsdk.ResultReason.Canceled:
    #     cancellation_details = result.cancellation_details
    #     print("Speech Recognition canceled: {}".format(cancellation_details.reason))
    #     if cancellation_details.reason == speechsdk.CancellationReason.Error:
    #         print("Error details: {}".format(cancellation_details.error_details))
    #         print("Did you set the speech resource key and region values?")

    return all_results
```

```python
texts = []

for i in tqdm(v_name):
  sound = AudioSegment.from_file(i)
  sound.export('wav_audio.wav', format='wav')

  text = from_file()
  texts.append(text)
```

```python
vf_df = pd.DataFrame({'text':texts})
vf_df['text'] = vf_df['text'].apply(lambda x: ' '.join(x))
vf_df.head(3)
```

```python
summary_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/FlyAI/project/summarization/summarization_train.csv')
summary_df = summary_df.sample(frac=1)
summary_df = summary_df.drop(['summary'], axis=1).head(600)
summary_df.head(3)
```

```python
vf = pd.concat([vf_df, summary_df])
vf.head(3)
```
